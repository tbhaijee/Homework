```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
sc_data = pd.read_csv(school_data_to_load)
st_data = pd.read_csv(student_data_to_load)
```

```python
per_st_budget = sc_data['budget'].divide(sc_data['size'])
```

```python
sc_data['per_st_budget'] = per_st_budget
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
sc_to = len(sc_data)
st_to = len(st_data)
sc_bud_to = '${:,.2f}'.format(sc_data.budget.sum())
avg_math = round(st_data['math_score'].mean(),2)
avg_read = round(st_data['reading_score'].mean(),2)
per_pass_math = round(((st_data['math_score'][st_data['math_score'] >=70].count())/st_to)*100,2)
per_pass_read = round(((st_data['reading_score'][st_data['reading_score'] >=70].count())/st_to)*100,2)
per_pass_over =round((per_pass_math + per_pass_read ) /2 ,2)
```

```python
sc_sum = pd.DataFrame({'Total Schools':sc_to,
                      'Total Students':st_to,
                      'Total Budget':sc_bud_to,
                      'Math Average':avg_math,
                      'Reading Average':avg_read,
                      '% Passing Math':per_pass_math,
                      '% Passing Reading':per_pass_read,
                      '% Passing Overall':per_pass_over},index=[0])
sc_sum
```

```python
# Combine the data into a single dataset
sc_data_comp= pd.merge(st_data, sc_data, how="left", on=["school_name", "school_name"])
sc_data_comp.head(3)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
sc_sum = sc_data_comp.groupby(['school_name','type'])
sc_sum
```

```python
sc_sum_st = sc_sum['student_name'].count()
sc_sum_bud = sc_sum['per_st_budget'].sum()
sc_sum_bud_f = sc_sum_bud.map('${:,.2f}'.format)
sc_sum_bud_per = (sc_sum_bud / sc_sum_st)
sc_sum_bud_per_f = sc_sum_bud_per.map('${:,.2f}'.format)
av_math = round(sc_sum['math_score'].mean(),2)
av_read = round(sc_sum['reading_score'].mean(),2)
```

```python
bins = [0,69.9,100]
gp = ["Fail","Pass"]
sc_data_comp["math"]=pd.cut(sc_data_comp["math_score"],bins=bins,labels=gp,include_lowest=True)
sc_data_comp["read"]=pd.cut(sc_data_comp["reading_score"],bins=bins,labels=gp,include_lowest=True)
```

```python
sc_sum
```

```python
pass_math = sc_data_comp.loc[sc_data_comp['math'] == 'Pass',:]
group_math = pass_math.groupby(['school_name','type'])
pass_read = sc_data_comp.loc[sc_data_comp['read'] == 'Pass',:]
group_read = pass_read.groupby(['school_name','type'])
```

```python
per_math=group_math['math_score'].count()
per_pass_math = round((per_math/sc_sum_st)*100,2)
per_read=group_read['reading_score'].count()
per_pass_read = round((per_read/sc_sum_st)*100,2)
per_pass_overall = (per_pass_math +per_pass_read)/2
```

```python
sc_summary = pd.DataFrame({'Students':sc_sum_st,
                          'Budget':sc_sum_bud_f,
                          'Per Student Budget':sc_sum_bud_per_f,
                          'Math Average':av_math,
                          'Reading Average':av_read,
                          '% Passing Math':per_pass_math,
                          '% Passing Reading':per_pass_read,
                          '% Passing Overall':per_pass_overall})
sc_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top_five = sc_summary.sort_values('% Passing Overall', ascending = False)
top_five.head(5)
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
top_five = sc_summary.sort_values('% Passing Overall', ascending = True)
top_five.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
grade9 = sc_data_comp[sc_data_comp['grade'] == "9th"]
grade10 = sc_data_comp[sc_data_comp['grade'] == "10th"]
grade11 = sc_data_comp[sc_data_comp['grade'] == "11th"]
grade12 = sc_data_comp[sc_data_comp['grade'] == "12th"]
gp_grade9 = grade9.groupby(['school_name'])
gp_grade10 = grade10.groupby(['school_name'])
gp_grade11 = grade11.groupby(['school_name'])
gp_grade12 = grade12.groupby(['school_name'])
avg_math_9 = round(gp_grade9["math_score"].mean())
avg_math_10 = round(gp_grade10["math_score"].mean())
avg_math_11 = round(gp_grade11["math_score"].mean())
avg_math_12 = round(gp_grade12["math_score"].mean())
math_scores_by_grade = pd.DataFrame({'9th':avg_math_9,
                                    '10th':avg_math_10,
                                    '11th':avg_math_11,
                                    '12th':avg_math_12})

math_scores_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending = [0, 585, 615, 645, 675]
gp_names = ["$0-585", "$585-615", "$615-645", "$645-675"]
```

```python
sc_data_comp["per_st_budget"]=pd.cut(sc_data_comp["per_st_budget"],bins=spending,labels=gp_names,include_lowest=True)
```

```python
spending= sc_data_comp.groupby('per_st_budget')
spending
```

```python
avg_math = round(spending['math_score'].mean(),2)
avg_read = round(spending['reading_score'].mean(),2)
sp_st= spending['student_name'].count()
```

```python
sp_pass_math = sc_data_comp.loc[sc_data_comp['math'] == 'Pass',:]
sp_math = sp_pass_math.groupby('per_st_budget')
sp_pass_read = sc_data_comp.loc[sc_data_comp['read'] == 'Pass',:]
sp_read = sp_pass_read.groupby('per_st_budget')
```

```python
per_math=sp_math['math_score'].count()
sp_per_pass_math = round((per_math/sp_st)*100,2)
per_read=sp_read['reading_score'].count()
sp_per_pass_read = round((per_read/sp_st)*100,2)
sp_per_pass_overall = (sp_per_pass_math +sp_per_pass_read)/2
```

```python
scores_by_sc_spending = pd.DataFrame({'Math Average':avg_math,
                          'Reading Average':avg_read,
                          '% Passing Math':sp_per_pass_math,
                          '% Passing Reading':sp_per_pass_read,
                          '% Passing Overall':sp_per_pass_overall})
scores_by_sc_spending 
```
