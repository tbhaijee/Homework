```python
# Add dependencies
import glob
import pandas as pd
```

```python
# add all the monthly data together

trip_data = pd.concat([pd.read_csv(f) for f in glob.glob('Resources/Monthly_Data_2019/*.csv')], ignore_index = True)
```

```python
# add all the quarerly data together

quarterly_trip_data = pd.concat([pd.read_csv(f) for f in glob.glob('Resources/Quarterly_Data_2019/*.csv')], ignore_index = True)
```

```python
trip_data.gender.replace([0, 1, 2], ['Unknown','Male', 'Female'], inplace=True)
trip_data.head()
```

```python
# Export Data

trip_data.to_csv("Resources/City_Bike_Data_2019/city_bike_data_2019.csv")
```
